```python
%matplotlib widget
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import numpy as np

#help(np.linspace)
xline = np.linspace(-5, 5, 101)
yline = xline
#print(xline)

X, Y = np.meshgrid(xline, yline)
print(X)
print(Y)

def sigmoid(x):
    sig = 1 / (1 + np.exp(-x))
    return sig - 0.5 # align to x-axis for analysis later

def linear(x):
    return x

def act(x1, x2, act_func):
    w1 = 5
    w2 = -3
    b = 0
    input_val = w1 * x1 + w2 * x2 + b
    activated = act_func(input_val)
    return activated

def get_color(Z):
    Z_clr = np.empty(Z.shape, dtype=str)
    Z_clr[Z>=0] = 1
    Z_clr[Z<0] = 0
    return Z_clr


Z_linear = act(X, Y, linear)
Z_sigmoid = act(X, Y, sigmoid)
```

```python
# https://stackoverflow.com/questions/332289/how-do-you-change-the-size-of-figures-drawn-with-matplotlib
plt.rcParams["figure.figsize"] = (12,12)

from matplotlib import cm
```

```python
fig1 = plt.figure()
plt.contour(X, Y, Z_linear)
plt.scatter(X, Y, c=get_color(Z_linear))

fig2, ax2 = plt.subplots(subplot_kw={"projection": "3d"})
surf2 = ax2.plot_surface(X, Y, Z_linear, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
# Add a color bar which maps values to colors.
#fig2.colorbar(surf2, shrink=0.5, aspect=5)
```

```python
fig3 = plt.figure()
plt.contour(X, Y, Z_sigmoid)
plt.scatter(X, Y, c=get_color(Z_sigmoid))

fig4, ax4 = plt.subplots(subplot_kw={"projection": "3d"})
surf4 = ax4.plot_surface(X, Y, Z_sigmoid, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
# Add a color bar which maps values to colors.
#fig4.colorbar(surf4, shrink=0.5, aspect=5)
```
